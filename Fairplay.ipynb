# Code for creating leg table

this code is triggered elsewhere.

Many of the values in the legTable are taken from the torm leg table.

The exception is the distance calculations which are calculated using geopy, if there are valid numbers to calculate. 



```python
def createLegTable(JourneyDF, posDF, portsDF, LegTable, indexCount):
    count = 0
    for index, row in JourneyDF.iterrows():
        count += 1
        coordDist = 0
        preLat = None
        preLong = None

        startDate = row["journey_start_timestamp"]
        endDate = row["journey_end_timestamp"]
        ImoNr = row["ship_id"]

        #imoNr + port
        LegTable.at[index, "ImoNr"] = row["ship_id"]
        LegTable.at[index, "Departure_Port"] = row["journey_start_location_name"] #
        LegTable.at[index, "Destination_Port"] = row["journey_end_location_name"]

        #Time
        LegTable.at[index, "journey_start_timestamp"] = startDate
        LegTable.at[index, "Min. ReportClosure-updated"] = endDate

        dfDepaPorts = portsDF.loc[portsDF["Ports"] == row["journey_start_location_name"]]
        dfDestPorts = portsDF.loc[portsDF["Ports"] == row["journey_end_location_name"]]

        depaPortLat = 0
        depaPortLong = 0

        destPortLat = 0
        destPortLong = 0
        
        try:
            LegTable.at[index, "RegionDepa"] = dfDepaPorts["Gruppe"].values.item(-1)
            depaPortLat = float(dfDepaPorts["Latitude"].values.item(-1) )
            LegTable.at[index, "Depature_port_latitude"] = str(depaPortLat).replace(".", ",")
            depaPortLong = float(dfDepaPorts["Longtitude"].values.item(-1) )
            LegTable.at[index, "Depature_port_longtitude"] = str(depaPortLong).replace(".", ",")
        except: 
            LegTable.at[index, "RegionDepa"] = 404
            LegTable.at[index, "Depature_port_latitude"] = 404
            LegTable.at[index, "Depature_port_longtitude"] = 404
        try:
            LegTable.at[index, "RegionDest"] = dfDestPorts["Gruppe"].values.item(-1)
            destPortLat = dfDestPorts["Latitude"].values.item(-1)
            LegTable.at[index, "Destination_port_latitude"] = str(destPortLat).replace(".", ",")
            destPortLong = dfDestPorts["Longtitude"].values.item(-1)
            LegTable.at[index, "Destination_port_longtitude"] = str(destPortLong).replace(".", ",")
        except: 
            LegTable.at[index, "RegionDest"] = 404
            LegTable.at[index, "Destination_port_latitude"] = 404
            LegTable.at[index, "Destination_port_longtitude"] = 404

     
        filtDF = posDF.loc[(posDF["ais_position_timestamp"].apply(pd.to_datetime) >= startDate) & (posDF["ais_position_timestamp"].apply(pd.to_datetime) <= endDate) & (posDF["ship_imo"] == ImoNr)]
        realCountOfIndex = 0
        realBestDest = 0
        realLat = 0
        realLong = 0
        realDist = 0
        countOfIndex = 0
        newEndTime = 0
      
        
        for filtIndex, filtRow in filtDF.iterrows():
            #newEndTime = 0
            #countOfIndex += 1
            
            lat = filtRow["ais_position_latitude"]
            long = filtRow["ais_position_longitude"]
            
            if type(lat) != float():
                lat = str(lat).replace(",", ".")
            if type(long) != float():
                long = str(long).replace(",", ".")
            anchorPoint = filtRow["nav_status_desc"]

            if preLat is not None:
                try:
                    dist = distance((lat, long), (preLat, preLong)).miles * 0.868976242
                    coordDist += round(dist, 2)
                except:
                    print("Failure here:")
                    print(lat, long, preLat, preLong)
                    preLat = 0
                    preLong = 0
            

            if anchorPoint == "At anchor" and filtIndex != 1 or anchorPoint == "Moored" and filtIndex != 1 or anchorPoint == "At anchor" and filtIndex != 0 or anchorPoint == "Moored" and filtIndex != 0:
                try:
                    bestDest = distance((lat, long), (destPortLat, destPortLong)).miles * 0.868976242
                except:
                    print("Fail here:")
                    print(lat, long, preLat, preLong)
                    bestDest = 9999
                

                if bestDest <= realBestDest and bestDest != 0:
                    realBestDest = bestDest
                    realLat = lat
                    realLong = long
                    realDist = coordDist
                    realCountOfIndex = countOfIndex
                    newEndTime = filtRow["ais_position_timestamp"]
                if bestDest == 9999:
                    pass
                elif realBestDest == 0:
                    realBestDest = bestDest
                    newEndTime = filtRow["ais_position_timestamp"]
                

            preLat = lat
            preLong = long
            #filtCount = int(countOfIndex)
        #LegTable.at[index, "Leg_id_legs (# of reports)"] = filtCount
        LegTable.at[index, "Leg_id_legs (# of reports)"] = filtDF["ais_position_longitude"].count()
        
        LegTable.at[index, "(closest anchor point to port) journey_end_timestamp"] = newEndTime

        #if realDist != 0:
         #   coordDist = realDist

        dfLegTable.at[index, "Coordinate_distance_sum"] = str(coordDist).replace(".", ",")

        #get days
        legLengthsDay = row["journey_underway_time_hours"]
        legLengthsDay = float(str(legLengthsDay).replace(",", ".")) 
        #print(float(legLengthsDay) )
        legLengthsDay = round(legLengthsDay / 24, 0)

        try:
            newEndTime = newEndTime.split(" ")[0]
            endDate = datetime.datetime.strptime(str(newEndTime), '%Y-%m-%d')
        except: pass


        daysTravelled = (endDate - startDate).days
        LegTable.at[index, "Leg Length (days covered)"] = daysTravelled
        #dfLegTable.at[index, "Leg Length (days covered)"] = row["journey_days"]
        LegTable.at[index, "Loading Condition"] = row["journey_laden"]

        #distance
        LegTable.at[index, "Logged_distance_sum"] = str(row["journey_underway_distance"]).replace(".", ",") #
        LegTable.at[index, "Avg. Speed-Median"] = str(row["journey_underway_calc_avg_speed"]).replace(".", ",")
        
        LegTable.at[index, "journey_id"] = row['journey_id']
        NoFilterDt = False
        try:
            firstRowFilt = filtDF.iloc[0]
            firstRowLat = str(firstRowFilt["ais_position_latitude"]).replace(",", ".")
            firstRowLong = str(firstRowFilt["ais_position_longitude"]).replace(",", ".")
        except:
            NoFilterDt = True


        try:
            if realDist != 0:
                lastRowLat = realLat
                lastRowLong = realLong
            else:
                lastRowFilt = filtDF.iloc[-1]
                lastRowLat = str(lastRowFilt["ais_position_latitude"]).replace(",", ".")
                lastRowLong = str(lastRowFilt["ais_position_longitude"]).replace(",", ".")
        except: 
            pass
        
        if NoFilterDt == False:
            firstRowDist = 0
            
            try:
                if depaPortLat != 404 and depaPortLong != 404 and depaPortLat != 0:
                    firstRowDist = distance((depaPortLat, depaPortLong), (firstRowLat, firstRowLong)).miles * 0.868976242
                    coordDist += firstRowDist
                else:
                    firstRowDist = 404
            except:
                firstRowDist = 404

            lastRowDist = 0
            try: 
                if destPortLat != 404 and destPortLong != 404 and destPortLat != 0:
                    lastRowDist = distance((destPortLat, destPortLong), (lastRowLat, lastRowLong)).miles * 0.868976242
                    coordDist += lastRowDist
                else:
                    lastRowDist = 404
            except: 
                lastRowDist = 404
                
            LegTable.at[index, "Distance to departure port"] = str(firstRowDist).replace(".", ",")
            LegTable.at[index, "Distance to destination port"] = str(lastRowDist).replace(".", ",")
        else:
            pass
        
        LegTable.at[index, "Journey table start Draught"] = row["journey_start_draft"]
        LegTable.at[index, "Journey table start Draught"] = row["journey_end_draft"]
        LegTable.at[index, "Journey table start Draught"] = row["journey_start_draft"]
        
        if count%500 == 0:
            LegTable.to_csv("over 3000 Leg Table.csv", sep = ";", index = False)
            print("=====================================")
            print(index)
        count += 1
    LegTable.to_csv("over 3000 Leg Table.csv", sep = ";", index = False)
    return LegTable
```

# Trigger the legtable creator

```python
"""
Nyt leg table skal have

ImoNr; Leg_id_legs (# of reports); Leg Length (days covered); Avg. Speed-Median; Avg. speed by time weight; 
std of Speed-Median; # of ME-Power reports; # of valid ME-Power reports; Min_Max_Draft_Normalization;
Loading Condition; Min. ReportClosure-updated; Departure_Port; Destination_Port; DistanceToPilotStation;
Observed_distance_sum; Logged_distance_sum; Coordinate_distance_sum; Median_distance_sum;
Depature_port_latitude ;Depature_port_longtitude ;Destination_port_latitude ;Destination_port_longtitude ;
Distance to departure port ;Distance to destination port ;RegionDepa ;RegionDest
"""


"""
columns = The columns wanted in the new leg table

dfJourney = sorted dfJourney imported from above, this value is later filtered to only include the journeys over 3000nautical miles
dfPos = sorted dfPos imported from above

lenPos = length of position

shipID = unique ship ID

BASIS logic in for loop.

for every unique shipID:
    filter position table on shipID
    filter journey table on shipID
    
    if either of the filtered tables are not empty:
        Using the dates and shipID to take a subset of the torm "leg table"
        trigger leg table creation code, this returns a leg table of the shipID
"""


columns = ["journey_id","ImoNr", "ship type", "Leg_id_legs (# of reports)", "Leg Length (days covered)", "Avg. Speed-Median", "Avg. speed by time weight", 
"std of Speed-Median", "# of ME-Power reports", "# of valid ME-Power reports", "Journey table start Draught", "Journey table end Draught", "Min_Max_Draft_Normalization",
"Loading Condition", "Departure_Port", "Destination_Port", "DistanceToPilotStation",
"Observed_distance_sum", "Logged_distance_sum", "Coordinate_distance_sum", "Median_distance_sum",
"Depature_port_latitude", "Depature_port_longtitude", "Destination_port_latitude", "Destination_port_longtitude",
"Distance to departure port", "Distance to destination port", "RegionDepa", "RegionDest",
"journey_start_timestamp", "Min. ReportClosure-updated", "(closest anchor point to port) journey_end_timestamp"]

dfLegTable = pd.DataFrame(columns = columns)

#dfLegTable = pd.read_csv("Leg Table.csv")

dfJourney = dfJourney.sort_values(['journey_start_timestamp'])
dfPos = dfPos.sort_values(['ais_position_timestamp'])



lenPos = len(dfPos)
dfJourney = dfJourney[dfJourney["journey_distance_nm"] >= 3000]



shipID = dfJourney["ship_id"].unique()

countIm = 0
for i in shipID:
    start = 0
    end = 7000
    dfNewPos = dfPos.loc[(dfPos["ship_imo"] == i)]
    #print(dfNewPos)
    dfNewJourney = dfJourney.loc[(dfJourney["ship_id"] == i)]
    print("dfNewJourney", len(dfNewJourney))
    lenPos = len(dfNewPos)
    print('lenPos', lenPos)
    done = False
    print("I = ", i)
    countIm += 1
    print("number of imo", countIm)
    if dfNewPos.empty == True or dfNewJourney.empty == True:
        pass
    else:
        while not done:

            dfTempPos = dfNewPos[start : end]
            df = dfTempPos
            df[["ais_position_timestamp"]] = dfTempPos[["ais_position_timestamp"]].apply(pd.to_datetime)
            df = df.sort_values(['ais_position_timestamp'])
            startDate = df["ais_position_timestamp"].iloc[0]
            endDate = df["ais_position_timestamp"].iloc[-1]

            print(len(df))

            dfTestTable = dfNewJourney.loc[(dfNewJourney["journey_start_timestamp"] >= startDate) & (dfNewJourney["journey_end_timestamp"].apply(pd.to_datetime) <= endDate)]

            indexCount = dfLegTable["ImoNr"].count()
            indexCount = 0

            dfLegTable = createLegTable(dfTestTable, df, dfCorrPorts, dfLegTable, indexCount)

            if end > lenPos:
                dfLegTable = createLegTable(dfTestTable, df, dfCorrPorts, dfLegTable, indexCount)
                dfLegTable.to_csv("over 3000 Leg Table.csv", sep = ";",  index = False)
                done = True
            else:
                start += 5000
                end += 5000
                percentage = (end / lenPos) * 100
                print("end", end, "percentage", percentage)
        print("===============")

    
```
